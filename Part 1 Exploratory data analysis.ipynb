### 1. Import Data and Creating 15 Minute Intervals

```python
import pandas as pd
import json
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_json('logins.json')
df.head()
```

```python
df.info()
```

```python
## Creating 15 minute intervals

df.set_index('login_time', inplace=True)
df['count'] = 1
df_resampled = df.resample('15T', label='right').sum()
df_resampled.head()
```

```python
df_resampled.info()
```

<a id='section_2'></a>
### 2. Plotting All Data


```python
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(df_resampled.index, df_resampled['count'])
plt.title('Plot of Login Count in 15 minute intervals')
plt.xlabel('Login Times')
plt.ylabel('Count');
```

<div class="alert alert-block alert-warning">

Its clear that there is some underlying pattern, but it is challenging to see in the plot. Let's use a rolling average plot to get a clearer picture of the pattern.
    
</div>

```python
df_resampled.rolling(window=50).mean().plot(figsize=(12, 8))
plt.title('Rolling Plot of Login Average in 15 minute intervals')
plt.xlabel('Login Times')
plt.ylabel('Count Average');
```

<div class="alert alert-block alert-success">

This plot gives a much clearer sense of the pattern. Its seems like there is a weekly cycle, with a spike in a few days of the weel regularly. Lets now take a closer look at each week by month to understand which days have those spikes in counts.
    
</div>

<a id='section_3'></a>
### 3. Plotting Data By Month

```python
plt.rcParams['figure.figsize'] = (12, 14)
fig, axes = plt.subplots(3, 1) 

df_1 = df_resampled.loc['1970-01']
axes[0].plot(df_1.index, df_1['count'])
axes[0].set_title('January Plot of Login Count in 15 minute intervals')
axes[0].set_ylabel('Count')

df_2 = df_resampled.loc['1970-02']
axes[1].plot(df_2.index, df_2['count'])
axes[1].set_title('February Plot of Login Count in 15 minute intervals')
axes[1].set_ylabel('Count')

df_3 = df_resampled.loc['1970-03']
axes[2].plot(df_3.index, df_3['count'])
axes[2].set_title('March Plot of Login Count in 15 minute intervals')
axes[2].set_ylabel('Count')
```

<div class="alert alert-block alert-warning">

Just as before, its a little unclear in these plots where the pattern lies. Let's use another rolling average plot to get a clearer view.    
</div>

```python
fig, axes = plt.subplots(3, 1) 

axes[0].plot(df_1.rolling(window=50).mean())
axes[0].set_title('January Rolling Plot of Login Average in 15 minute intervals')
axes[0].set_ylabel('Count Average')

axes[1].plot(df_2.rolling(window=50).mean())
axes[1].set_title('February Rolling Plot of Login Average in 15 minute intervals')
axes[1].set_ylabel('Count Average')

axes[2].plot(df_3.rolling(window=50).mean())
axes[2].set_title('March Rolling Plot of Login Average in 15 minute intervals')
axes[2].set_ylabel('Count Average');
```

<div class="alert alert-block alert-success">

This plot gives a much clearer sense of the pattern. It seems that there are peaks for a few days then they drop again. My thought would be that the peaks are on the weekends. Let's take a closer look at the weeks. 
</div>

<a id='section_4'></a>
### 4. Plotting Data By Week

```python
datetime(1970, 3, 1).weekday()
```

Since 3-1-1970 is a Sunday, we will start our plots on 3-2-1970, a Monday, to keep all of the weekend together at the end of our plot. So our plots below are Monday-Sunday.

```python
fig, axes = plt.subplots(4, 1) 

df_1 = df_resampled.loc['1970-03-02' : '1970-03-08']
axes[0].plot(df_1.index, df_1['count'])
axes[0].set_title('Week 1 Plot of Login Count in 15 minute intervals')
axes[0].set_ylabel('Count')

df_2 = df_resampled.loc['1970-03-09': '1970-03-15']
axes[1].plot(df_2.index, df_2['count'])
axes[1].set_title('Week 2 Plot of Login Count in 15 minute intervals')
axes[1].set_ylabel('Count')

df_3 = df_resampled.loc['1970-03-16': '1970-03-22']
axes[2].plot(df_3.index, df_3['count'])
axes[2].set_title('Week 3 Plot of Login Count in 15 minute intervals')
axes[2].set_ylabel('Count')

df_4 = df_resampled.loc['1970-03-23': '1970-03-29']
axes[3].plot(df_4.index, df_4['count'])
axes[3].set_title('Week 4 Plot of Login Count in 15 minute intervals')
axes[3].set_ylabel('Count')
```

<div class="alert alert-block alert-warning">

The peaks do seem to be higher on the weeken, with the exception of Saint Patrick's Day. But let's use a rolling average plot to get a clearer picture of the pattern.
    
</div>

```python
fig, axes = plt.subplots(4, 1) 

axes[0].plot(df_1.rolling(window=80).mean())
axes[0].set_title('Week 1 Rolling Plot of Login Average in 15 minute intervals')
axes[0].set_ylabel('Count Average')

axes[1].plot(df_2.rolling(window=80).mean())
axes[1].set_title('Week 2 Rolling Plot of Login Average in 15 minute intervals')
axes[1].set_ylabel('Count Average')

axes[2].plot(df_3.rolling(window=80).mean())
axes[2].set_title('Week 3 Rolling Plot of Login Average in 15 minute intervals')
axes[2].set_ylabel('Count Average')

axes[3].plot(df_4.rolling(window=80).mean())
axes[3].set_title('Week 4 Rolling Plot of Login Average in 15 minute intervals')
axes[3].set_ylabel('Count Average');
```

<div class="alert alert-block alert-success">

With these plots, we can see a lot clearer that the weekend has much higher peaks than the week. The only exception is week 3, which again has a holiday, so that is less accurate of standard trends. 
</div>

### 5. Plotting Data By Day

```python
plt.rcParams['figure.figsize'] = (12, 30)
fig, axes = plt.subplots(7, 1) 

df_1 = df_resampled.loc['1970-03-02']
axes[0].plot(df_1.index, df_1['count'])
axes[0].set_title('3/2 Plot of Login Count in 15 minute intervals')
axes[0].set_ylabel('Count')

df_2 = df_resampled.loc['1970-03-03']
axes[1].plot(df_2.index, df_2['count'])
axes[1].set_title('3/3 Plot of Login Count in 15 minute intervals')
axes[1].set_ylabel('Count')

df_3 = df_resampled.loc['1970-03-04']
axes[2].plot(df_3.index, df_3['count'])
axes[2].set_title('3/4 Plot of Login Count in 15 minute intervals')
axes[2].set_ylabel('Count')

df_4 = df_resampled.loc['1970-03-05']
axes[3].plot(df_4.index, df_3['count'])
axes[3].set_title('3/5 Plot of Login Count in 15 minute intervals')
axes[3].set_ylabel('Count')

df_5 = df_resampled.loc['1970-03-06']
axes[4].plot(df_5.index, df_3['count'])
axes[4].set_title('3/6 Plot of Login Count in 15 minute intervals')
axes[4].set_ylabel('Count')

df_6 = df_resampled.loc['1970-03-07']
axes[5].plot(df_6.index, df_3['count'])
axes[5].set_title('3/7 Plot of Login Count in 15 minute intervals')
axes[5].set_ylabel('Count')

df_7 = df_resampled.loc['1970-03-08']
axes[6].plot(df_7.index, df_3['count'])
axes[6].set_title('3/8 Plot of Login Count in 15 minute intervals')
axes[6].set_ylabel('Count')
```

<div class="alert alert-block alert-success">

When looking at the pattern by day, the peak hours seem to be at 12 o'clock in the afternoon and midnight. It seems to be consistant for all of the days in this week. 
</div>

### 6. Summary

Over all, it seems that there is a general pattern for the time that users are using the service. They are more frequently using the service on the weekends over the weekdays. And on any day, the peak hours for use are at 12 o'clock PM in the afternoon and at 12 o'clock AM, midnight.
